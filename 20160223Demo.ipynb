```python
from bs4 import BeautifulSoup as bs
soup = bs(res.text)
domain = 'http://www.appledaily.com.tw'
list_item = soup.select('.rtddt a')
for li in list_item:
    print li.select('h1')[0].text, li.select('h2')[0].text, li.select('time')[0].text, domain + li['href']
```

```python
import pandas as pd

domain = 'http://www.appledaily.com.tw'
ary = []
for li in list_item:
    ary.append({'title':li.select('h1')[0].text, \
                'category':li.select('h2')[0].text, \
                'time':li.select('time')[0].text, \
                'url':domain + li['href']})
```

```python
print ary[0]['title']
```

```python
df = pd.DataFrame(ary)
```
